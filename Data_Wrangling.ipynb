```python
# We gathered electricity usage data 2023 from extracting and split data for each participant based on the Meter ID 
# and then combine the data from all 12 sheets into a single DataFrame for each participant.


# Load the Excel file
file_path = "directory/UD_Case_Study2023.xlsx"

# Read all sheet names
sheet_names = pd.ExcelFile(file_path).sheet_names

# Dictionary to store combined DataFrame for each participant
cityOfNewark_participants_data = {}

# Iterate through each sheet
for sheet_name in sheet_names:
    # Read the current sheet into a DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    
    # Ensure the 'Meter ID' column exists
    if 'Meter ID' not in df.columns:
        raise ValueError(f"'Meter ID' column not found in sheet: {sheet_name}")
    
    # Group by 'Meter ID' and append data for each participant
    for participant_id, participant_df in df.groupby('Meter ID'):
        # If the participant is already in the dictionary, append the data
        if participant_id in cityOfNewark_participants_data:
            cityOfNewark_participants_data[participant_id] = pd.concat(
                [cityOfNewark_participants_data[participant_id], participant_df], 
                ignore_index=True
            )
        else:
            # If the participant is new, add their data to the dictionary
            cityOfNewark_participants_data[participant_id] = participant_df

# Save each participant's combined data to a CSV file
for participant_id, participant_df in cityOfNewark_participants_data.items():
    output_file = f"electric_usage_{participant_id}.csv"
    participant_df.to_csv(output_file, index=False)
    print(f"Saved data for participant {participant_id} to {output_file}")
```

#### Green Button Data

**Loads and processes electric usage data from a main folder and its subfolders. Combines files into a single DataFrame per folder prefix.**

```python
# We gathered electricity usage data from the each Households power consumption of the N:18 for last 2 years

def find_data_start(file_path, header_keyword="TYPE"):
    """
    Find the starting row for actual data based on a keyword in the file.
    Returns the row number to use with `skiprows`.
    """
    with open(file_path, 'r') as file:
        for i, line in enumerate(file):
            if header_keyword in line:
                return i  # Return the row index where data starts
    return None  # Return None if keyword is not found


def load_electric_usage_data(main_folder_path):
    # Dictionary to hold each DataFrame with a unique key
    data_frames = {}
    
    # Search for CSV files that start with 'electric_usage' in all subfolders of main folder
    file_pattern = os.path.join(main_folder_path, '**', 'electric_usage*.csv')
    file_paths = glob.glob(file_pattern, recursive=True)
    
    # Dictionary to track files by their folder prefix (e.g., P1, P2)
    folder_files = {}
    
    # Group files by folder prefix (e.g., P1, P2)
    for file in file_paths:
        folder_prefix = os.path.basename(os.path.dirname(file))
        
        if folder_prefix.startswith("P") and folder_prefix[1:].isdigit():
            folder_number = int(folder_prefix[1:])
            if 1 <= folder_number <= 18:
                if folder_prefix not in folder_files:
                    folder_files[folder_prefix] = []
                folder_files[folder_prefix].append(file)
    
    # Process each folder's files, creating a single DataFrame per folder prefix
    for folder_prefix, files in folder_files.items():
        data_list = []
        for f in files:
            try:
                # Determine whether to use skiprows
                if folder_prefix in ['P6', 'P8']or (folder_prefix.startswith('P') and folder_prefix[1:].isdigit() and 12 <= int(folder_prefix[1:]) <= 18):
                    # For P6 and P8, start from the first row without skipping
                    df = pd.read_csv(f, encoding='utf-8', on_bad_lines='skip', low_memory=False)
                    print(f"Loaded {os.path.basename(f)} successfully, starting from the first row for {folder_prefix}.")
                else:
                    # For other folders, determine the start row based on keyword
                    start_row = find_data_start(f)
                    if start_row is None:
                        print(f"Data header row not found in {f}, skipping file.")
                        continue
                    
                    # Read the file starting from the identified header row
                    df = pd.read_csv(f, encoding='utf-8', skiprows=start_row, on_bad_lines='skip', low_memory=False)
                    print(f"Loaded {os.path.basename(f)} successfully, starting from row {start_row + 1} for {folder_prefix}.")
                
                data_list.append(df)
            except pd.errors.ParserError as e:
                print(f"Error reading {f}: {e}")
                continue  # Skip problematic files

        # Concatenate all files in a folder into a single DataFrame if multiple exist
        if data_list:
            combined_df = pd.concat(data_list, ignore_index=True)
            # Assign the DataFrame a standardized name based on the folder prefix
            standardized_name = f"{folder_prefix}_electric_usage_df"
            globals()[standardized_name] = combined_df
            data_frames[standardized_name] = combined_df
            print(f"{standardized_name} created successfully.")
        else:
            print(f"Warning: No valid data for {folder_prefix}")

    return data_frames

# # Loading Usage data 
main_folder = "directory/Data" 
electric_usage_data_frames = load_electric_usage_data(main_folder)
```

```python
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta
```

```python
def parse_date(date_str):
    """Convert XML timestamp to datetime object."""
    return datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S %z').replace(tzinfo=None)
```

```python
def process_export_xml(file_path, cutoff_start_date, cutoff_end_date):
    """
    Parse an export.xml file and extract steps, heart rate, calories, and sleep data.
    Returns four DataFrames: df_steps, df_heart_rate, df_calories, and df_sleep.
    """
    # Initialize lists to collect data
    steps_data = []
    heart_rate_data = []
    calories_data = []
    sleep_data = []

    # Parse the XML file
    tree = ET.parse(file_path)
    root = tree.getroot()

    # Parse each record in the XML
    for record in root.findall("Record"):
        start_date = parse_date(record.get('startDate'))
        
        # Filter records based on date range
        if start_date < cutoff_start_date or start_date > cutoff_end_date:
            continue  # Skip records outside the last two years
        
        # Steps Data
        if record.get('type') == 'HKQuantityTypeIdentifierStepCount':
            steps_data.append({
                'timestamp': start_date,
                'Steps': int(record.get('value'))
            })

        # Heart Rate Data
        elif record.get('type') == 'HKQuantityTypeIdentifierHeartRate':
            heart_rate_data.append({
                'timestamp': start_date,
                'HeartRate': float(record.get('value'))
            })

        # Caloric Burn Data
        elif record.get('type') in ['HKQuantityTypeIdentifierActiveEnergyBurned', 'HKQuantityTypeIdentifierBasalEnergyBurned']:
            calories_data.append({
                'timestamp': start_date,
                'CaloricBurn': float(record.get('value')),
                'type': 'active' if record.get('type') == 'HKQuantityTypeIdentifierActiveEnergyBurned' else 'basal'
            })
        
        # Sleep Data
        elif record.get('type') == 'HKCategoryTypeIdentifierSleepAnalysis':
            end_date = parse_date(record.get('endDate'))
            sleep_data.append({
                'startDate': start_date,
                'endDate': end_date
            })

    # Convert lists to DataFrames
    df_steps = pd.DataFrame(steps_data)
    df_heart_rate = pd.DataFrame(heart_rate_data)
    df_calories = pd.DataFrame(calories_data)
    df_sleep = pd.DataFrame(sleep_data)
    
    # Process Sleep Data if not empty
    if not df_sleep.empty:
        df_sleep['time_asleep'] = df_sleep['endDate'] - df_sleep['startDate']
        df_sleep = df_sleep.groupby(df_sleep['startDate'].dt.date).agg(
            total_time_asleep=('time_asleep', 'sum'),
            bed_time=('startDate', 'min'), 
            awake_time=('endDate', 'max'), 
            sleep_counts=('startDate', 'count'), 
            rem_cycles=pd.NamedAgg(column='time_asleep', aggfunc=lambda x: (x // timedelta(minutes=90)).sum())
        )
        df_sleep['time_in_bed'] = df_sleep['awake_time'] - df_sleep['bed_time']
        df_sleep['restless_time'] = df_sleep['time_in_bed'] - df_sleep['total_time_asleep']
        df_sleep['time_in_bed'] = df_sleep['time_in_bed'].dt.total_seconds() / 60  # Convert to minutes
        df_sleep['total_time_asleep'] = df_sleep['total_time_asleep'].dt.total_seconds() / 60  # Convert to minutes

    return df_steps, df_heart_rate, df_calories, df_sleep

def load_and_process_folders(main_folder_path, folders):
    """
    Process each specified folder, loading and parsing the 'export.xml' file within
    each 'apple_health_export-p#' subfolder.

    Returns a dictionary where each folder (P1, P4, etc.) has its own set of DataFrames.
    """
    all_data = {}

    # Define date range for the last two years
    cutoff_end_date = datetime.now()
    cutoff_start_date = cutoff_end_date - timedelta(days=365*2)
    
    for folder_prefix in folders:
        # Construct the path to the export.xml file within the apple_health_export-p# folder
        subfolder_name = f"apple_health_export-{folder_prefix.lower()}"
        file_path = os.path.join(main_folder_path, folder_prefix, subfolder_name, 'export.xml')
        
        # Check if the export.xml file exists
        if not os.path.isfile(file_path):
            print(f"File {file_path} not found. Skipping...")
            continue

        # Process the XML file and create DataFrames
        df_steps, df_heart_rate, df_calories, df_sleep = process_export_xml(file_path, cutoff_start_date, cutoff_end_date)

        # Store each DataFrame in a dictionary for the current folder
        all_data[folder_prefix] = {
            'steps_data': df_steps,
            'heart_rate_data': df_heart_rate,
            'calories_data': df_calories,
            'sleep_data': df_sleep
        }

        print(f"Data processing completed for folder {folder_prefix}.")

    return all_data

# Define the main folder path and list of folders to process
main_folder = "directory"
folders_to_process = ['P1', 'P4', 'P5', 'P7', 'P9', 'P10', 'P11', 'P12', 'P14', 'P15', 'P17', 'P18']
apple_health_data = load_and_process_folders(main_folder, folders_to_process)
```

```python
def create_individual_dataframes(apple_health_data, folders, data_types):
    """
    Creates individual DataFrames for each specified folder and data type in the
    apple_health_data dictionary, with consistent naming.
    
    Parameters:
    - apple_health_data: Dictionary containing data for each folder.
    - folders: List of folder names to process (e.g., ['P1', 'P4']).
    - data_types: List of data types to create DataFrames for (e.g., ['steps_data', 'heart_rate_data']).
    """
    for folder in folders:
        if folder in apple_health_data:
            for data_type in data_types:
                if data_type in apple_health_data[folder]:
                    # Create a standardized name for each DataFrame
                    df_name = f"{folder}_{data_type}_df"
                    
                    # Assign the DataFrame a standardized name based on its folder prefix
                    globals()[df_name] = apple_health_data[folder][data_type]
                    
                    # Print confirmation
                    print(f"Created DataFrame: {df_name}")
                else:
                    print(f"{data_type} not available for {folder}")

# Usage
folders_to_create = ['P1', 'P4', 'P5', 'P7', 'P9', 'P10', 'P11', 'P12', 'P14', 'P15', 'P17', 'P18']
data_types_to_create = ['steps_data', 'heart_rate_data', 'calories_data', 'sleep_data']

create_individual_dataframes(apple_health_data, folders_to_create, data_types_to_create)

# Now you can access each DataFrame by its standardized name, e.g., P1_steps_data_df, P4_heart_rate_data_df, etc.
```

```python
import json


def load_fitbit_data(base_dir, folders):
    """
    Loads steps, heart rate, calories, and sleep data from Fitbit files located in specified subfolders.

    Parameters:
    - base_dir: The base directory where the main folders (e.g., P3, P8, P16) are located.
    - folders: List of folder names to process (e.g., ['P3', 'P8', 'P16']).

    Each folder is expected to contain:
    - For most folders:
        - Steps and Calories: Monthly CSV files in TIMESERIES_GoogleData subfolder.
        - Heart Rate: Daily CSV files in TIMESERIES_GoogleData subfolder.
        - Sleep: JSON files in Global Export Data subfolder.
    - For P16:
        - All data (Steps, Heart Rate, Calories, Sleep) are JSON files located in Global Export Data subfolder.

    Returns:
    - DataFrames are created as global variables for each folder and data type.
    """
    for folder in folders:
        # Define the path to the Global Export Data subfolder
        global_export_path = os.path.join(base_dir, folder, 'Takeout', 'Fitbit', 'Global Export Data')

        # Define JSON file patterns for each data type
        data_types = {
            'steps': 'steps-*.json',
            'calories': 'calories-*.json',
            'heart_rate': 'heart_rate-*.json',
            'sleep': 'sleep-*.json'
        }

        # Process each data type
        for data_type, pattern in data_types.items():
            # Find all matching JSON files for the current data type
            json_files = glob.glob(os.path.join(global_export_path, pattern))
            data_records = []

            # Load data from all JSON files
            for json_file in json_files:
                try:
                    with open(json_file, 'r') as f:
                        data = json.load(f)
                        data_records.extend(data)  # Combine data from multiple files
                except Exception as e:
                    print(f"Error loading {json_file} for {folder} ({data_type}): {e}")
            
            # Convert the combined data to a DataFrame
            if data_records:
                globals()[f"{folder}_{data_type}_data_df"] = pd.DataFrame(data_records)
                print(f"Loaded {data_type} data for {folder} from {len(json_files)} files in {global_export_path}")
            else:
                globals()[f"{folder}_{data_type}_data_df"] = pd.DataFrame()
                print(f"No {data_type} files found for {folder} in {global_export_path}")


# Usage
base_directory = 'directory'
folders_to_load = ['P3', 'P8', 'P16']

# Load data from Fitbit folders within P3 and P8
load_fitbit_data(base_directory, folders_to_load)
```

```python
def display_head_of_data(folders):
    """
    Displays the first four rows of each data type (steps, heart_rate, calories, sleep) 
    for each specified folder (e.g., P3, P8).
    
    Parameters:
    - folders: List of folder names (e.g., ['P3', 'P8']).
    """
    data_types = ['steps_data', 'heart_rate_data', 'calories_data', 'sleep_data']
    
    for folder in folders:
        print(f"--- Data for {folder} ---\n")
        for data_type in data_types:
            df_name = f"{folder}_{data_type}_df"
            
            # Check if the DataFrame exists in the global scope
            if df_name in globals():
                df = globals()[df_name]
                
                # Display the first 4 rows if the DataFrame is not empty
                if not df.empty:
                    print(f"{data_type} for {folder}:")
                    print(df.head(4))
                    print("\n")
                else:
                    print(f"{data_type} for {folder} is empty.\n")
            else:
                print(f"{data_type} for {folder} not found.\n")
        print("="*50)

# Usage example
folders_to_display = ['P3', 'P8', 'P16']
display_head_of_data(folders_to_display)
```

```python
# Constants for calculations
WALKING_STRIDE_LENGTH = 0.75  # meters
WALKING_SPM = 100            # Steps per minute

# Define a function to calculate steps
def calculate_steps(row):
    if not pd.isnull(row['Distance (meters)']):  # If distance data is available
        return row['Distance (meters)'] / WALKING_STRIDE_LENGTH
    elif not pd.isnull(row['Duration (min)']):  # Fallback to duration-based estimate
        return row['Duration (min)'] * WALKING_SPM
    else:
        return None  # Return None if neither Distance nor Duration is available

# Apply the function to each row and add a new column for steps
df_workoutData['Steps'] = df_workoutData.apply(calculate_steps, axis=1)

# Display the DataFrame with the new column
print(df_workoutData.head(5))
```

```python
data_rows_with_ranges = {
    'startDate': ['Aug 27 - Sep 2', 'Jul 30 - Aug 5', 'Jun 25 - Jul 1', 'May 28 - Jun 3',
                  'Apr 30 - May 6', 'Mar 26 - Apr 1', 'Feb 27 - Mar 4', 'Jan 30 - Feb 5',
                  '12/26/2024 - Jan 1', 'Nov 28 - Dec 4', 'Oct 31 - Nov 6'],
    'total_time_asleep': ['5h 46min', '7h 6min', '5h 43min', '5h 48min', '5h 39min', '6h 10min', 
                    '6h 4min', '6h 19min', '6h 32min', '7h 8min', '4h 48min'],
    'bed_time': ['1:26 AM', '12:18 AM', '1:10 AM', '12:43 AM', '12:05 AM', '1:06 AM',
                 '1:35 AM', '12:40 AM', '12:55 AM', '12:29 AM', '2:08 AM'],
    'awake_time': ['7:36 AM', '8:33 AM', '7:11 AM', '7:02 AM', '6:32 AM', '7:38 AM', 
                   '7:58 AM', '7:38 AM', '8:10 AM', '8:09 AM', '7:04 AM']
}
df = pd.DataFrame(data_rows_with_ranges)

# Function to expand date ranges into individual rows
def expand_date_ranges(df, date_col):
    expanded_rows = []
    
    for _, row in df.iterrows():
        if ' - ' in row[date_col]:  # Identify ranges
            # Split the range into start and end
            start_str, end_str = row[date_col].split(' - ')
            
            # Handle year information for cross-year ranges
            if '/' in start_str or '/' in end_str:
                start_date = pd.to_datetime(start_str, errors='coerce')
                end_date = pd.to_datetime(end_str, errors='coerce')
            else:
                year = "2024"  # Use default year or adjust dynamically
                start_date = pd.to_datetime(f"{start_str} {year}", format='%b %d %Y', errors='coerce')
                end_date = pd.to_datetime(f"{end_str} {year}", format='%b %d %Y', errors='coerce')
            
            # Generate dates for the range
            if not pd.isna(start_date) and not pd.isna(end_date):
                date_range = pd.date_range(start_date, end_date)
                
                # Replicate the row for each date in the range
                for date in date_range:
                    new_row = row.copy()
                    new_row[date_col] = date
                    expanded_rows.append(new_row)
        else:
            # Keep rows without ranges as-is
            expanded_rows.append(row)
    
    # Create a new DataFrame from the expanded rows
    return pd.DataFrame(expanded_rows)

# Expand the ranges in 'startDate'
expanded_df = expand_date_ranges(df, 'startDate')

# Display the resulting DataFrame
print(expanded_df)
```

```python
def compare_sleep_and_usage_for_day(sleep_folders, usage_folders, year=2024, month=9, day=21):
    """
    Compares sleep durations with electricity usage over time for a specific day.

    Parameters:
    - sleep_folders: List of folder names containing sleep data (e.g., ['P1', 'P2', ..., 'P17']).
    - usage_folders: List of folder names containing electricity usage data (e.g., ['P1', 'P2', ..., 'P17']).
    - year: The year to filter data (e.g., 2024).
    - month: The month to filter data (e.g., 9 for September).
    - day: The day to filter data (e.g., 20).
    """
    for sleep_folder, usage_folder in zip(sleep_folders, usage_folders):
        # Construct the names of the DataFrames
        sleep_df_name = f"{sleep_folder}_sleep_data_df_clean"
        usage_df_name = f"{usage_folder}_electric_usage_df_clean"

        if sleep_df_name in globals() and usage_df_name in globals():
            sleep_data = globals()[sleep_df_name]
            usage_data = globals()[usage_df_name]

            # Ensure timestamps are datetime
            sleep_data['bed_time'] = pd.to_datetime(sleep_data['bed_time'], errors='coerce')
            usage_data['timestamp'] = pd.to_datetime(usage_data['timestamp'], errors='coerce')

            # Filter for the specific day
            sleep_data = sleep_data[
                (sleep_data['bed_time'].dt.year == year) &
                (sleep_data['bed_time'].dt.month == month) &
                (sleep_data['bed_time'].dt.day == day)
            ]
            usage_data = usage_data[
                (usage_data['timestamp'].dt.year == year) &
                (usage_data['timestamp'].dt.month == month) &
                (usage_data['timestamp'].dt.day == day)
            ]

            if not sleep_data.empty and not usage_data.empty:
                # Aggregate sleep data by hour
                sleep_data = sleep_data.set_index('bed_time')
                sleep_aggregated = sleep_data[['total_time_asleep']].resample('h').mean()
                usage_data = usage_data.set_index('timestamp')
                usage_aggregated = usage_data[['USAGE_kWh']].resample('h').sum()

                # Ensure hourly range is complete
                full_range = pd.date_range(start=f"{year}-{month:02d}-{day:02d} 00:00", 
                                           end=f"{year}-{month:02d}-{day:02d} 23:59", freq='h')
                sleep_aggregated = sleep_aggregated.reindex(full_range).fillna(0)
                usage_aggregated = usage_aggregated.reindex(full_range).fillna(0)

                # Merge datasets
                combined_data = pd.merge(
                    sleep_aggregated, usage_aggregated, left_index=True, right_index=True, how='inner'
                )

                if not combined_data.empty:
                    # Plot the data
                    fig, ax1 = plt.subplots(figsize=(12, 6))

                    ax1.plot(combined_data.index, combined_data['total_time_asleep'], color='b', label='Total Time Asleep (minutes)')
                    ax1.set_xlabel('Hour')
                    ax1.set_ylabel('Total Time Asleep (minutes)', color='b')
                    ax1.tick_params(axis='y', labelcolor='b')

                    ax2 = ax1.twinx()
                    ax2.plot(combined_data.index, combined_data['USAGE_kWh'], color='r', label='Electricity Usage (kWh)')
                    ax2.set_ylabel('Electricity Usage (kWh)', color='r')
                    ax2.tick_params(axis='y', labelcolor='r')

                    ax1.xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%H:%M'))
                    ax1.set_xlim([pd.Timestamp(year, month, day), pd.Timestamp(year, month, day, 23, 59)])

                    plt.title(f'Sleep Durations vs Electricity Usage for {sleep_folder} on {year}-{month:02d}-{day:02d}')
                    ax1.legend(loc='upper left')
                    ax2.legend(loc='upper right')
                    plt.tight_layout()
                    plt.show()
                else:
                    print(f"No overlapping data for {sleep_folder} and {usage_folder} on {year}-{month:02d}-{day:02d}.")
            else:
                print(f"No data available for {sleep_folder} or {usage_folder} on {year}-{month:02d}-{day:02d}.")
        else:
            print(f"No data available for {sleep_folder} or {usage_folder}.")


# Usage for 20 September 2024 for P1 to P17
participants_to_compare = [f'P{i}' for i in range(1, 18)]
compare_sleep_and_usage_for_day(participants_to_compare, participants_to_compare, year=2023, month=9, day=29)
```

```python
# Function to aggregate minute-level data into hourly
def aggregate_to_hourly(df, value_col, aggregation='sum'):
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df.set_index('timestamp', inplace=True)
    if aggregation == 'mean':  # For HeartRate
        hourly_data = df[value_col].resample('h').mean()
    else:  # For Steps and Calories
        hourly_data = df[value_col].resample('h').sum()
    hourly_data = hourly_data.reset_index()
    return hourly_data

# Function to merge data with electricity usage
def merge_with_electricity(data_df, electricity_df, value_col):
    electricity_df['timestamp'] = pd.to_datetime(electricity_df['timestamp'])
    merged_df = pd.merge(data_df, electricity_df[['timestamp', 'USAGE_kWh']], on='timestamp', how='inner')
    merged_df = merged_df.rename(columns={value_col: 'Value'})
    return merged_df

# Metric-specific configurations
metrics = {
    "HeartRate": {"df_prefix": "P#_heart_rate_data_df_clean", "column": "HeartRate", "aggregation": "mean"},
    "Calories": {"df_prefix": "P#_calories_data_df_clean", "column": "CaloricBurn", "aggregation": "sum"},
    "Steps": {"df_prefix": "P#_steps_data_df_clean", "column": "Steps", "aggregation": "sum"}
}

# Loop through metrics and participants
for metric, details in metrics.items():
    for i in range(1, 18):  # Loop from P1 to P17
        # Define data frame names
        activity_df_name = details['df_prefix'].replace('#', str(i))
        electricity_df_name = f'P{i}_electric_usage_df_clean'
        value_col = details['column']
        aggregation = details['aggregation']

        try:
            # Assume the dataframes are available in the namespace
            activity_df = globals()[activity_df_name]
            electricity_df = globals()[electricity_df_name]

            # Aggregate activity data to hourly
            hourly_data = aggregate_to_hourly(activity_df, value_col, aggregation)

            # Merge with electricity usage data
            merged_data = merge_with_electricity(hourly_data, electricity_df, value_col)

            # Generate scatter plot
            plt.figure(figsize=(8, 6))
            plt.scatter(merged_data['Value'], merged_data['USAGE_kWh'], alpha=0.6)
            plt.title(f'P{i} - {metric} vs. Electricity Usage')
            plt.xlabel(metric)
            plt.ylabel('Electricity Usage (kWh)')
            plt.grid(True)
            plt.show()

        except KeyError as e:
            print(f"Dataframe {activity_df_name} or {electricity_df_name} not found: {e}")
        except Exception as ex:
            print(f"An error occurred for participant P{i} with {metric}: {ex}")
```
